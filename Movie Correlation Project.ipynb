```python
# Importing Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use("ggplot")
from matplotlib import figure

%matplotlib inline
# matplotlib.RcParams[figure.figsize] = (12,8)
```

```python
# Data source
# https://www.kaggle.com/datasets/danielgrijalvas/movies/data
# Read Data

df = pd.read_csv("movies.csv")
df.head()
```

```python
# Checking Missing Data
for col in df.columns:
    percentage_missing = np.mean(df[col].isna())
    print(f"{col} = {percentage_missing:.2}%")
    
# Removing NULL data
df = df.dropna()
```

```python

# Changing data type of from float to int
# Removing .0 from float
df["budget"] = df["budget"].astype("int64")
df["gross"] = df["gross"].astype("int64")
df["score"] = df["score"].astype("int64")
df["votes"] = df["votes"].astype("int64")
df["runtime"] = df["runtime"].astype("int64")

df.dtypes
```

```python
# Creating new correct year because some of "year" and "released" data don't match
df["correct_year"] = df["released"].str.extract(pat="([0-9]{4})").astype(int)
df.head()
```

```python
# Sorting from biggest gross
df.sort_values(by=["gross"], inplace = False, ascending = False)
```

```python
# Drop any duplicates
df["company"].drop_duplicates().sort_values(ascending=False)
```

```python
# Scatter plot Budget vs gross
plt.scatter(x=df["budget"], y=df["gross"])
plt.title("Budget vs Gross earning")
plt.xlabel("Budget")
plt.ylabel("Gross")
plt.show()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x="budget", y="gross", data = df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Calculatin correlation

numeric_df = df.select_dtypes(include=["float64", "int64"])
numeric_df.corr()
```

```python
# High correlation between budget and gross
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movies Feature")
plt.ylabel("Movies Feature")
plt.show()
```

```python
# Changing non-cumerical features into categorical

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == "object"):
        df_numerized[col_name] = df_numerized[col_name].astype("category")
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
# Correlation Matrix for all features
correlation_matrix = df_numerized.corr()
plt.figure(figsize= (12, 8))
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movies Feature")
plt.ylabel("Movies Feature")
plt.show()
```

```python
# Changing correlation matrix model for readable purposes
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()

sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```
